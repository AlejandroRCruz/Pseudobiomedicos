# Calidad del agua en México (2012–2024)
## Equipo: Pseudobiomédicos
**Fuente de base de datos:** CONAGUA — Indicadores de Calidad del Agua Subterránea https://www.gob.mx/conagua/es/articulos/indicadores-de-calidad-del-agua?idiom=es

**Objetivo:** Analizar la calidad del agua, en específico contaminantes que fomenten el sobre peso /obesidad en México. Relación ODS 6 y ODS 3.

## Descripción de la base de datos
De acuerdo a la fuente original, el archivo XLSX cuenta con 14 parámetros: Fluoruros (FLUORUROS), Coliformes fecales (COLI_FEC), Nitrógeno de Nitratos (N_NO3), Arsénico Total (AS_TOT), Cadmio Total (CD_TOT), Cromo Total (CR_TOT), Mercurio Total (HG_TOT), Plomo Total (PB_TOT), Alcalinidad (ALC), Conductividad eléctrica (CONDUCT), Dureza Total (DUR), Sólidos Disueltos Totales (SDT), Hierro Total (FE_TOT) y Manganeso Total (MN_TOT). Estos abarcan diferentes estados de la República Mexicana en un periodo de 2012 a 2014

## Impacto en la salud
- La documentación investigada dice que la exposición crónica al arsénico en agua potable ha sido vinculado con resistencia a la insulina y mayor riesgo de diabetes tipo 2, condiciones estrechamente relacionadas con obesidad y síndrome metabólico (Navas-Acien et al., 2006; Paul et al., 2011).
- El exceso de fluoruros puede alterar la función tiroidea y el metabolismo basal, favoreciendo disfunciones metabólicas (Peckham & Awofeso, 2014).
- La exposición al plomo interfiere con el metabolismo de lípidos y proteínas, y ha sido asociado con el síndrome metabólico en adultos mexicanos (Téllez-Rojo et al., 2019; Wang et al., 2020).
- El cadmio, al acumularse en riñón e hígado, genera estrés oxidativo y disfunciones metabólicas que predisponen a obesidad y diabetes (Satarug et al., 2010; Xu et al., 2021).
- El mercurio actúa como disruptor endocrino, alterando el metabolismo energético y aumentando el riesgo de obesidad y resistencia a la insulina (Bjørklund et al., 2017).
- El cromo, aunque es esencial,  en exceso puede ser tóxico; su suplementación ha mostrado efectos en el control de glucosa y peso en personas con sobrepeso (Vincent, 2011; Balk et al., 2007).
- El manganeso en altas concentraciones altera la homeostasis metabólica y el metabolismo energético, con evidencia emergente de su relación con disfunciones metabólicas (Chen et al., 2018).

## Umbrales utilizados
**Umbrales utilizados:**
- **Arsénico:** > 0.01 mg/L (límite OMS; más conservador que el mexicano de 0.025 mg/L). La literatura muestra efectos metabólicos desde este umbral.
- **Fluoruros:** ≥ 1.5 mg/L (CONAGUA "Alta"; límite OMS y NOM-127).
- **Plomo:** > 0.01 mg/L (CONAGUA "No apta FAAP").
- **Cadmio:** > 0.003 mg/L (CONAGUA "Apta FAAP" — umbral intermedio con efectos sobre insulina).
- **Mercurio:** > 0.006 mg/L (CONAGUA "No apta FAAP").
- **Cromo:** > 0.05 mg/L (CONAGUA "No apta FAAP").
- **Manganeso:** > 0.4 mg/L (CONAGUA "Puede afectar la salud").

## Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Estética global
plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'font.size': 11,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.grid': True,
    'grid.alpha': 0.25,
    'grid.linestyle': '--',
    'figure.dpi': 130,
    'axes.titlesize': 13,
    'axes.titleweight': 'bold',
})

# Paleta de colores por contaminante
COLORES = {
    'AS_TOT':    '#E24B4A',   # Arsénico — rojo
    'FLUORUROS': '#EF9F27',   # Fluoruros — ámbar
    'PB_TOT':    '#7F77DD',   # Plomo — púrpura
    'CD_TOT':    '#1D9E75',   # Cadmio — verde
    'HG_TOT':    '#D4537E',   # Mercurio — rosa
    'CR_TOT':    '#888780',   # Cromo — gris
    'MN_TOT':    '#378ADD',   # Manganeso — azul
}

NOMBRES = {
    'AS_TOT':    'Arsénico',
    'FLUORUROS': 'Fluoruros',
    'PB_TOT':    'Plomo',
    'CD_TOT':    'Cadmio',
    'HG_TOT':    'Mercurio',
    'CR_TOT':    'Cromo',
    'MN_TOT':    'Manganeso',
}
```

## Carga y exploración inicial de la base de datos

```python
# Ruta del archivo
RUTA_ARCHIVO = '/content/Calidad_del_Agua_Subterr_nea_p_2012-2024_15082025.xlsx'

df_raw = pd.read_excel(RUTA_ARCHIVO, engine='openpyxl')

print(f'Filas: {len(df_raw):,}')
print(f'Columnas: {df_raw.shape[1]}')
print('\nPrimeras columnas:')
print(df_raw.columns.tolist()[:10])
```

```python
# Vista rápida de la estructura
df_raw.head(3)
```

## Limpieza y preparación

```python
# Estandarizar nombres de columnas (quitar espacios, mayúsculas)
df_raw.columns = df_raw.columns.str.strip().str.upper()

# Columnas de los contaminantes elegidos y la información relevante
COLS_ID  = ['CLAVE SITIO', 'ESTADO', 'MUNICIPIO', 'ACUIFERO',
            'LATITUD', 'LONGITUD', 'PERIODO']
COLS_OBS = ['AS_TOT_MG/L', 'FLUORUROS_MG/L', 'PB_TOT_MG/L',
            'CD_TOT_MG/L', 'HG_TOT_MG/L', 'CR_TOT_MG/L', 'MN_TOT_MG/L']
COLS_CAL = ['CALIDAD_AS', 'CALIDAD_FLUO', 'CALIDAD_PB',
            'CALIDAD_CD', 'CALIDAD_HG', 'CALIDAD_CR', 'CALIDAD_MN']

# Verificar que existan las columnas esperadas
presentes = [c for c in COLS_OBS + COLS_CAL if c in df_raw.columns]
faltantes = [c for c in COLS_OBS + COLS_CAL if c not in df_raw.columns]
print('Columnas encontradas:', presentes)
if faltantes:
    print('\nATENCIÓN — columnas NO encontradas (revisar nombre exacto):', faltantes)
```

```python
# Si el PERIODO viene como string '2012-2024' o similar, extraemos el año de la medición
# Si el archivo tiene una fila por año, la columna PERIODO tendrá valores numéricos (2012, 2013...)
# Ajusta según lo que veas en df_raw

col_periodo = 'PERIODO' if 'PERIODO' in df_raw.columns else None
if col_periodo:
    df_raw['PERIODO'] = df_raw['PERIODO'].astype(str)
    # Si el formato es '2012-2024', usamos solo el primer año como referencia
    df_raw['AÑO'] = df_raw['PERIODO'].str.extract(r'(\d{4})').astype(float)
    print('Años encontrados:', sorted(df_raw['AÑO'].dropna().unique()))

# Seleccionar y copiar columnas relevantes
cols_disponibles = [c for c in COLS_ID + COLS_OBS + COLS_CAL + ['SEMÁFORO', 'CONTAMINANTES', 'AÑO']
                    if c in df_raw.columns]
df = df_raw[cols_disponibles].copy()

# Renombrar columnas de valores para simplificar
rename_map = {
    'AS_TOT_MG/L':    'AS_TOT',
    'FLUORUROS_MG/L': 'FLUORUROS',
    'PB_TOT_MG/L':    'PB_TOT',
    'CD_TOT_MG/L':    'CD_TOT',
    'HG_TOT_MG/L':    'HG_TOT',
    'CR_TOT_MG/L':    'CR_TOT',
    'MN_TOT_MG/L':    'MN_TOT',
}
df.rename(columns=rename_map, inplace=True)

print(f'DataFrame de trabajo: {df.shape}')
df.head(2)
```

## Gráficas

A continuación mostramos el top 15 de estados de la república mexicana.
Usamos los umbrales del diccionario de CONAGUA. Un sitio se marca como riesgo obesogénico en un contaminante si su valor supera el umbral OMS/CONAGUA indicado en la literatura.

```python
# Umbrales de alerta obesogénica (mg/L)
# Para arsénico usamos 0.01 (límite OMS) — más conservador que el mexicano de 0.025
UMBRALES = {
    'AS_TOT':    0.01,    # OMS; CONAGUA marca como "Apta FAAP" desde aquí
    'FLUORUROS': 1.5,     # CONAGUA "Alta" — altera microbiota
    'PB_TOT':    0.01,    # CONAGUA "No apta FAAP"
    'CD_TOT':    0.003,   # CONAGUA "Apta FAAP" — umbral intermedio
    'HG_TOT':    0.006,   # CONAGUA "No apta FAAP"
    'CR_TOT':    0.05,    # CONAGUA "No apta FAAP"
    'MN_TOT':    0.4,     # CONAGUA "Puede afectar la salud"
}

def limpiar_valor(val):
    """Convierte '<0.001' o '0.005' a float. NaN si no se puede."""
    if pd.isna(val):
        return np.nan
    s = str(val).strip().replace('<', '').replace('>', '').replace(',', '.')
    try:
        return float(s)
    except:
        return np.nan

# Limpiar columnas numericas
for col in UMBRALES.keys():
    if col in df.columns:
        df[col] = df[col].apply(limpiar_valor)

# columnas booleanas de alerta por contaminante
for col, umbral in UMBRALES.items():
    if col in df.columns:
        df[f'ALERTA_{col}'] = df[col] > umbral

# Columna agregada para saber si tiene al menos un obesógeno fuera de norma
alertas = [f'ALERTA_{c}' for c in UMBRALES if f'ALERTA_{c}' in df.columns]
df['OBESOGENO_PRESENTE'] = df[alertas].any(axis=1)

print(f'Sitios totales: {len(df):,}')
print(f'Sitios con al menos un obesógeno fuera de norma: {df["OBESOGENO_PRESENTE"].sum():,} ({df["OBESOGENO_PRESENTE"].mean()*100:.1f}%)')
```

## Sitios con obesógenos fuera de norma por estado

```python
# Conteo de sitios con alerta por estado y contaminante
df_estado = df.groupby('ESTADO')[[f'ALERTA_{c}' for c in UMBRALES if f'ALERTA_{c}' in df.columns]].sum()
df_estado.columns = [NOMBRES.get(c.replace('ALERTA_', ''), c.replace('ALERTA_', '')) for c in df_estado.columns]

# Ordenar por total descendente
df_estado['TOTAL'] = df_estado.sum(axis=1)
df_estado = df_estado.sort_values('TOTAL', ascending=False).drop(columns='TOTAL')

# Tomar top 15 estados
df_plot = df_estado.head(15)

fig, ax = plt.subplots(figsize=(14, 6))

bottom = np.zeros(len(df_plot))
for contam in df_plot.columns:
    orig_key = [k for k, v in NOMBRES.items() if v == contam]
    color = COLORES.get(orig_key[0], '#888780') if orig_key else '#888780'
    ax.bar(df_plot.index, df_plot[contam], bottom=bottom, label=contam,
           color=color, width=0.7, edgecolor='none')
    bottom += df_plot[contam].values

ax.set_title('Sitios con obesógenos fuera de norma por estado (2012–2024)',
             pad=12)
ax.set_ylabel('Número de sitios')
ax.set_xlabel('')
ax.set_xticklabels(df_plot.index, rotation=35, ha='right', fontsize=10)
ax.legend(loc='upper right', fontsize=9, framealpha=0.5,
          title='Contaminante', title_fontsize=9)
ax.yaxis.set_major_locator(mticker.MaxNLocator(integer=True))

fig.tight_layout()
fig.savefig('grafica1_barras_estado.png', bbox_inches='tight', dpi=150)
plt.show()
print('Guardada: grafica1_barras_estado.png')
```

## Prevalencia nacional de cada obesógeno

```python
# % de sitios fuera de norma por tipo de contaminante
prevalencia = {}
for col in UMBRALES:
    alerta_col = f'ALERTA_{col}'
    if alerta_col in df.columns:
        total = df[col].notna().sum()
        if total > 0:
            pct = df[alerta_col].sum() / total * 100
            prevalencia[NOMBRES[col]] = round(pct, 1)

prev_series = pd.Series(prevalencia).sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(8, 5))

colores_plot = []
for nombre in prev_series.index:
    key = [k for k, v in NOMBRES.items() if v == nombre]
    colores_plot.append(COLORES.get(key[0], '#888780') if key else '#888780')

bars = ax.barh(prev_series.index, prev_series.values,
               color=colores_plot, height=0.6, edgecolor='none')

for bar, val in zip(bars, prev_series.values):
    ax.text(val + 0.3, bar.get_y() + bar.get_height()/2,
            f'{val}%', va='center', fontsize=10)

ax.set_title('Prevalencia nacional de cada obesógeno\n(% de sitios fuera de norma)', pad=12)
ax.set_xlabel('% de sitios monitoreados')
ax.xaxis.set_major_formatter(mticker.PercentFormatter())
ax.set_xlim(0, prev_series.max() * 1.2)

fig.tight_layout()
fig.savefig('grafica2_prevalencia_nacional.png', bbox_inches='tight', dpi=150)
plt.show()
print('Guardada: grafica2_prevalencia_nacional.png')
```

## Co-ocurrencia arsénico + fluoruros por estado

```python
# Sitios donde AMBOS están fuera de norma simultáneamente
if 'ALERTA_AS_TOT' in df.columns and 'ALERTA_FLUORUROS' in df.columns:
    df['CO_AS_FLUO'] = df['ALERTA_AS_TOT'] & df['ALERTA_FLUORUROS']

    # % por estado sobre el total de sitios con datos en ambos parámetros
    co_estado = df.groupby('ESTADO').apply(
        lambda g: g['CO_AS_FLUO'].sum() / len(g) * 100
    ).sort_values(ascending=False).head(12)

    fig, ax = plt.subplots(figsize=(10, 5))

    ax.bar(co_estado.index, co_estado.values,
           color='#E24B4A', width=0.6, edgecolor='none', alpha=0.85)

    for i, (estado, val) in enumerate(co_estado.items()):
        ax.text(i, val + 0.3, f'{val:.1f}%', ha='center', fontsize=9)

    ax.set_title('Co-ocurrencia arsénico + fluoruros fuera de norma\n(% de sitios por estado)', pad=12)
    ax.set_ylabel('% de sitios con ambos obesógenos')
    ax.set_xticklabels(co_estado.index, rotation=35, ha='right', fontsize=10)
    ax.yaxis.set_major_formatter(mticker.PercentFormatter())

    fig.tight_layout()
    fig.savefig('grafica3_coocurrencia.png', bbox_inches='tight', dpi=150)
    plt.show()
    print('Guardada: grafica3_coocurrencia.png')
else:
    print('Columnas ALERTA_AS_TOT o ALERTA_FLUORUROS no encontradas. Revisa el paso de limpieza.')
```

## Índice de exposición: % sitios contaminados vs población estatal

```python
# Población estatal 2020 (INEGI) — millones de habitantes
POBLACION = {
    'AGUASCALIENTES': 1.43, 'BAJA CALIFORNIA': 3.77, 'BAJA CALIFORNIA SUR': 0.80,
    'CAMPECHE': 0.93, 'CHIAPAS': 5.54, 'CHIHUAHUA': 3.74, 'CIUDAD DE MÉXICO': 9.21,
    'COAHUILA DE ZARAGOZA': 3.15, 'COLIMA': 0.73, 'DURANGO': 1.83,
    'GUANAJUATO': 6.17, 'GUERRERO': 3.54, 'HIDALGO': 3.08, 'JALISCO': 8.35,
    'MÉXICO': 16.99, 'MICHOACÁN DE OCAMPO': 4.75, 'MORELOS': 1.97,
    'NAYARIT': 1.24, 'NUEVO LEÓN': 5.78, 'OAXACA': 4.13, 'PUEBLA': 6.58,
    'QUERÉTARO': 2.37, 'QUINTANA ROO': 1.86, 'SAN LUIS POTOSÍ': 2.82,
    'SINALOA': 3.03, 'SONORA': 2.94, 'TABASCO': 2.40, 'TAMAULIPAS': 3.64,
    'TLAXCALA': 1.34, 'VERACRUZ DE IGNACIO DE LA LLAVE': 8.06,
    'YUCATÁN': 2.32, 'ZACATECAS': 1.62,
}

# % de sitios con al menos un obesógeno por estado
pct_estado = df.groupby('ESTADO').apply(
    lambda g: g['OBESOGENO_PRESENTE'].sum() / len(g) * 100
).reset_index()
pct_estado.columns = ['ESTADO', 'PCT_OBESOGENOS']

# Número absoluto de sitios contaminados (para tamaño de burbuja)
n_estado = df.groupby('ESTADO')['OBESOGENO_PRESENTE'].sum().reset_index()
n_estado.columns = ['ESTADO', 'N_SITIOS_CONT']

burbuja = pct_estado.merge(n_estado, on='ESTADO')
burbuja['ESTADO_UPPER'] = burbuja['ESTADO'].str.upper().str.strip()
burbuja['POBLACION'] = burbuja['ESTADO_UPPER'].map(
    {k.upper(): v for k, v in POBLACION.items()}
).fillna(1.0)

fig, ax = plt.subplots(figsize=(11, 7))

scatter = ax.scatter(
    burbuja['PCT_OBESOGENOS'],
    burbuja['POBLACION'],
    s=burbuja['N_SITIOS_CONT'] * 12 + 40,
    c=burbuja['PCT_OBESOGENOS'],
    cmap='RdYlGn_r',
    alpha=0.75,
    edgecolors='none'
)

# Etiquetas de los estados más relevantes
top_estados = burbuja.nlargest(10, 'PCT_OBESOGENOS')
for _, row in top_estados.iterrows():
    nombre = row['ESTADO'].split()[0] if len(row['ESTADO'].split()) > 1 else row['ESTADO']
    ax.annotate(nombre, (row['PCT_OBESOGENOS'], row['POBLACION']),
                textcoords='offset points', xytext=(6, 4),
                fontsize=8.5, color='#3d3d3a')

ax.set_title('Índice de exposición obesogénica por estado\n(tamaño = número absoluto de sitios contaminados)', pad=12)
ax.set_xlabel('% de sitios con al menos un obesógeno fuera de norma')
ax.set_ylabel('Población estatal (millones, INEGI 2020)')
ax.xaxis.set_major_formatter(mticker.PercentFormatter())

plt.colorbar(scatter, ax=ax, label='% sitios contaminados', shrink=0.7)

fig.tight_layout()
fig.savefig('grafica5_burbujas_exposicion.png', bbox_inches='tight', dpi=150)
plt.show()
print('Guardada: grafica5_burbujas_exposicion.png')
```
